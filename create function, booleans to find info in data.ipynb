Q1: Create a function that checks each row in the 311 dataframe to see if the word "graffiti" is present in either the complaint_type value or descriptor value. Both columns may contain the word, so you should check both. If the word "graffiti" is found, the function should return the boolean value True. If "graffiti" is not found, the function should return the boolean value False.

```
def graffiti_counts(row):
  if pd.notnull(row.descriptor):  
    if 'graffiti' in row.complaint_type and row.complaint_type != 'NaN':
      return True
    elif 'graffiti' in row.descriptor and row.descriptor != 'NaN':
      return True
    else:
      return False
```

```
def graffiti_counts_py(row):
#Tried it a second time with nested IF statements before searching for 'graffiti'. Used above function for the rest of the assignment though. 
  if pd.notnull(row.descriptor):  
    if row.complaint_type != 'NaN':
      if row.descriptor != 'NaN':
        if 'graffiti' in row.complaint_type:
          return True
        elif 'graffiti' in row.descriptor:
          return True
        else:
          return False
```

```
df['graffiti_flag'] = df.apply(graffiti_counts, axis=1)
df.head(10)
```
